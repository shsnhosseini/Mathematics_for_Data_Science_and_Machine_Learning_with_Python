### Vectors

#### 1. Creating Vectors

```python
# Vector as a Python list
vector_list = [18, 19, 28]

# Same numbers as a 1D numpy array (not dimensionless - it has one dimension)
vector_array = np.array(vector_list)  # shape: (3,)

# Row vector (2D array with shape (1, 3))
row_vector = np.array([[18, 19, 28]])  # or vector_array.reshape(1, -1), auto-calculates second dimension

# Column vector (2D array with shape (3, 1))
column_vector = np.array([[18], [19], [28]])  # or vector_array.reshape(-1, 1), auto-calculates first dimension
```

```python
# Checking the sizes of the variables using shape method associated with numpy objects
print(np.shape(vector_list))
print(vector_array.shape)
print(row_vector.shape)
print(column_vector.shape)
```

```python
# For column/row vectors, you can also use np.expand_dims():
row_vector = np.expand_dims(vector_array, axis=0)
column_vector = np.expand_dims(vector_array, axis=1)
row_vector.shape, column_vector.shape
```

#### 2. Operations on Vectors

```python
# For simplicity the vectors are created orientationless.
# Creating vectors
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
```

```python
# Vector addition
v_add = v1 + v2 
print("Vector addition:", v_add)
```

```python
# Vector subtraction
v_sub = v1 - v2 
print("Vector subtraction:", v_sub)
```

```python
# Scalar multiplication
scalar = -2
v_scalar = scalar * v1
print("Scalar multiplication:", v_scalar)
```

```python
# Dot product (inner product)
dot_product = np.dot(v1, v2)  # 32 (1*4 + 2*5 + 3*6)
print("Dot product:", dot_product)
```

```python
# Cross product (only for 3D vectors)
# The cross product of two vectors, v1 and v2, is denoted as v1 × v2. 
# It produces a new vector, which is perpendicular to both v1 and v2. 
# The direction of the new vector is determined by the right-hand rule. 
cross_product = np.cross(v1, v2)
print("Cross product:", cross_product)
```

```python
# Outer product
# In older texts or informal contexts, the term "outer product" has been used to refer to the cross product,
# but in modern linear algebra, these are distinct concepts
# Outer product produces a matrix (or tensor) and is defined for any dimension;
# Cross product produces a vector and is only defined in three (or seven) dimensions
outer_product = np.outer(v1, v2)
print("Outer product:", outer_product)
```

```python
# Vector magnitude (norm)
# Mehod 1: Using numpy's built-in function
magnitude = np.linalg.norm(v1)
print("Magnitude of v1:", magnitude)
```

```python
# Method 2: using math library
math.sqrt(1**2 + 2**2 + 3**2) 
```

```python
# Unit vector
unit_v1 = v1 / np.linalg.norm(v1)
print("Unit vector of v1:", unit_v1)
print("Magnitude of unit vector:", np.linalg.norm(unit_v1))
```

```python
# Element-wise multiplication(Hadamard product)
hadamard = v1 * v2 
print("Hadamard product:", hadamard)
print(type(hadamard))
```

```python
# Angle between vectors (in radians)
cosine_theta = np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
angle = np.arccos(cosine_theta)
print("Angle in radians:", angle)

# Angle in degrees
# Method 1: Using math library
angle_degrees = math.degrees(angle)
print("Angle in degrees using math library:", angle_degrees)

# Method 2: Using numpy function
angle_degrees = np.rad2deg(angle)
print("Angle in degrees using numpy function:", angle_degrees)
```

```python
# Compute the angle in degrees directly in one line
angle_degrees = np.rad2deg(np.arccos(np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))))

print("Angle in degrees:", angle_degrees)
```

```python
# Projection of v1 onto v2
projection = (np.dot(v1, v2) / np.dot(v2, v2)) * v2
print(projection)
```

```python
# Vector rejection
rejection = v1 - projection
print(rejection)
```

```python
# Checking orthogonality
print(np.dot(v1, v2) == 0) #  Should be True if v1 and v2 are orthogonal
print(np.dot(v1, cross_product) == 0)  # Should be True if v1 and cross_product are orthogonal
print(np.dot(v2, cross_product) == 0)  # Should be True if v2 and cross_product are orthogonal
print(np.dot(rejection, projection) == 0)  # Should be True if projection and rejection are orthogonal
```

But, we know that vector projection and vector rejection are orthogonal to each other!

**Why This Happens:**

   - Computers represent real numbers with finite precision (e.g., 64-bit floating-point numbers in numpy).

   - Operations like projections, subtractions, and dot products accumulate small numerical errors.

   - Thus, `np.dot(rejection, projection)` yield a value like `-6e-16` instead of exactly `0`.

Thus, the issue you're encountering is due to **floating-point precision errors** in numerical computations. When you compute the dot product of the rejection and projection vectors, the result may not be exactly zero due to tiny rounding errors in floating-point arithmetic, even though **mathematically** they should be orthogonal.

```python
np.dot(rejection, projection)
```

**How to Fix It:**

   Instead of checking for exact equality `(== 0)`, you should check if the absolute value of the dot product is very close to zero (within a small tolerance). Use `np.isclose()` or check against a small epsilon value:

```python
# Option 1: Using np.isclose()
print(np.isclose(np.dot(rejection, projection), 0.0))  # True if nearly zero
```

```python
# Option 2: Using a Small Epsilon
epsilon = 1e-10
print(abs(np.dot(rejection, projection)) < epsilon)  # True if negligible
```
