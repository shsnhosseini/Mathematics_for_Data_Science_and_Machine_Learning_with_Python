$$ \textbf{Law of Large Numbers} $$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

*  This code is aexample of a **computational simulation** that demonstrates a theoretical law through random sampling. Specifically, it's a stochastic simulation.

*  The **Law of Large Numbers** states that as the sample size increases, the sample mean converges to the true population mean. This function visualizes that convergence.

```python
np.random.seed(42)  # for reproducibility

# Simulation of law of large numbers (LLN)
def law_large_number(distribution, n=100):
    """
    Simulate the law of large numbers for a given distribution.
    
    Parameters:
    distribution: scipy.stats distribution object (e.g., stats.norm, stats.expon)
    n: number of samples
    """
    # Get distribution name
    title = distribution.dist.name if hasattr(distribution, 'dist') else distribution.name
    
    # Generate random observations
    observations = distribution.rvs(size=n)
    
    # Calculate cumulative sample means
    sample_means = np.cumsum(observations) / np.arange(1, n+1)
    
    # Get theoretical mean
    mu = distribution.mean()
    
    # Create the plot
    fig, ax = plt.subplots(figsize=(8, 4))
    
    # Plot vertical lines from zero to each observation
    for i, obs in enumerate(observations, start=1):
        ax.plot([i, i], [0, obs], color='grey', alpha=0.5, linewidth=0.8)
    
    # Plot observation points
    ax.scatter(range(1, n+1), observations, color='grey', alpha=0.5, 
               s=30, label='_nolegend_')
    
    # Plot theoretical mean if it exists
    ax.axhline(y=mu, color='black', linewidth=1.5, linestyle='--', 
                label='Population Mean')
    
    # Plot cumulative sample means as a thick green line
    ax.plot(range(1, n+1), sample_means, linewidth=3, alpha=0.6, 
            color='green', label='Sample mean')
    
    # Set title and labels
    ax.set_title(title)
    ax.set_xlabel('Number of observations (n)')
    ax.set_ylabel('Value')
    ax.grid(True, alpha=0.3)
    ax.legend()
    
    return fig, ax
```

```python
# Example usage with Normal distribution
# Create normal distribution with mean=1, std=2
#normal_dist = stats.norm(loc=1, scale=2)
#fig, ax = law_large_number(normal_dist)
#plt.show()

# Example with Exponential distribution (rate parameter = 1/scale)
# For exponential with mean 2, scale=2
exponential_dist = stats.expon(scale=2)
fig, ax = law_large_number(exponential_dist)
plt.show()
```

This simulation (Monte Carlo Simulation) clearly shows how the green line eventually approaches the dashed black line as n increases, illustrating the Law of Large Numbers.

**Note That This is Used for Understanding, Not Real Data Analysis**

Rather than analyzing actual collected data, it:
*  Creates artificial data with known properties (we know the true mean)

*  Visually demonstrates a mathematical truth through simulated experimentation

*  Allows you to "see" the Law of Large Numbers in action
