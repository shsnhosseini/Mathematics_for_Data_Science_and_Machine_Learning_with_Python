$$ \textbf{Law of Large Numbers} $$

```python
using LinearAlgebra, Statistics, Random, Distributions
using Plots, LaTeXStrings
```

*  This code is aexample of a **computational simulation** that demonstrates a theoretical law through random sampling. Specifically, it's a stochastic simulation.

*  The **Law of Large Numbers** states that as the sample size increases, the sample mean converges to the true population mean. This function visualizes that convergence.

```python
# Simulation of law of large numbers(LLN)
function law_large_number(distribution, n=100)
    title = nameof(typeof(distribution))
    observations = rand(distribution, n)
    sample_means = cumsum(observations) ./ (1:n)
    μ = mean(distribution)
    plot(repeat((1:n)', 2),
        [zeros(1, n); observations'], label = "", color = :grey, alpha = 0.5)
    plot!(1:n, observations, color = :grey, markershape = :circle,
         alpha = 0.5, label = "", linewidth = 0)
    if !isnan(μ)
        hline!([μ], color = :black, linewidth = 1.5, linestyle = :dash,
                        grid = false, label = "Mean")
    end
    # cumulative sample means as a thick green line
    plot!(1:n, sample_means, linewidth = 3, alpha = 0.6, color = :green,
                     label = "Sample mean")
    return plot!(title = title)
end
```

```python
law_large_number(Normal(1, 2))
# law_large_number(Exponential(2)) # Exponential distribution
```

This simulation (Monte Carlo Simulation) clearly shows how the green line eventually approaches the dashed black line as n increases, illustrating the Law of Large Numbers.

**Note That This is Used for Understanding, Not Real Data Analysis**

Rather than analyzing actual collected data, it:
*  Creates artificial data with known properties (we know the true mean)

*  Visually demonstrates a mathematical truth through simulated experimentation

*  Allows you to "see" the Law of Large Numbers in action
